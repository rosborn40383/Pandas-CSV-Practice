```python
def convert(x):
    y=(x*(1.8)+32)
    return y
```

```python
df['temp_f']=list(map(lambda x: convert(x), df['temp_c']))
```

```python
df.sample(5)
```

```python
df=df[['date_time','temp_f','temp_c','relative_humidity','wind_speed','visibility','pressure','weather']]
```

```python
df.date_time=pd.to_datetime(df.date_time)
```

```python
months_date=df.date_time.dt.month
```

```python
def Month_Num(n):
    month_dict={1:'January', 2:'February',3:"March",4:'April',5:'May',6:'June',7:'July',8:'August',9:'September',10:'October',11:'November',12:'December'}
    return month_dict[n]
```

```python
Month_Num(7)
```

```python
monthly_avg.month=list(map(lambda x: Month_Num(x), monthly_avg['month']))
```

```python
def Hour_Num(n):
    hour_dict={0:"12:00AM",  1:"1:00AM",  2:"2:00AM",
             3:"3:00AM",  4:"4:00AM",  5:"5:00AM",
             6:"6:00AM",  7:"7:00AM",  8:"8:00AM",
             9:"9:00AM",  10:"10:00AM",  11:"11:00AM",
             12:"12:00PM",13:"1:00PM",  14:"2:00PM",
             15:"3:00PM",  16:"4:00PM",  17:"5:00PM",
             18:"6:00PM",  19:"7:00PM",  20:"8:00PM",
             21:"9:00PM",  22:"10:00PM",  23:"11:00PM"}
    return hour_dict[n]
```

```python
hourly_avg.index=list(map(lambda x: Hour_Num(x),hourly_avg.index))
```
